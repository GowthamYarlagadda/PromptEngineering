```python
import openai
import time
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation (optional, based on your requirements)
    text = "".join(char for char in text if char.isalnum() or char.isspace())

    # Tokenize the text into individual words
    words = text.split()

    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    words = [word for word in words if word not in stop_words]

    # Join the words back into a cleaned sentence
    cleaned_text = " ".join(words)

    return cleaned_text
```

```python
def trim_messages(messages, max_tokens):
    total_tokens = 0
    trimmed_messages = []

    for msg in messages:
        content_tokens = len(msg["content"].split())
        if total_tokens + content_tokens <= max_tokens:
            trimmed_messages.append(msg)
            total_tokens += content_tokens
        else:
            remaining_tokens = max_tokens - total_tokens
            trimmed_content = ' '.join(msg["content"].split()[:remaining_tokens])
            trimmed_msg = {"role": msg["role"], "content": trimmed_content}
            trimmed_messages.append(trimmed_msg)
            break

    return trimmed_messages
```

```python
def create_chatbot(system_msg, prompts_list):
    # Initialize messages with the initial system message
    messages = [{"role": "system", "content": clean_text(system_msg)}]

    for prompt in prompts_list:
        # Merge the system instruction into the user's message
        user_input = clean_text(f"{system_msg} {prompt}")
        messages.append({"role": "user", "content": user_input})

        # Check if the total tokens in messages exceed the model's limit
        total_tokens = sum(len(msg["content"].split()) for msg in messages)
        if total_tokens > 4096:
            # Trim messages to fit within the token limit
            messages = trim_messages(messages, max_tokens=4096)

        # Call OpenAI API to get the response
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=40,
            temperature=0.6
        )

        # Extract the reply from the API response
        reply = response["choices"][0]["message"]["content"]
        reply = clean_text(reply)

        # Append the assistant's reply to the messages
        messages.append({"role": "assistant", "content": reply})

        # Pause for a few seconds before the next prompt (optional)
        time.sleep(20)

        # Check if the user input is "quit()"
        if user_input == "quit()":
            break

    # Print the number of messages and inform that the chatbot is ready
    print(f"Total messages: {len(messages)}")
    print("Your new custom chatbot is ready!")
```
